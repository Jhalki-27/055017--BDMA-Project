Problem Statement:


The goal is to analyze movie-related data stored in MongoDB to uncover key trends, identify high-performing genres, top-rated movies, language preferences, and production patterns. Stakeholders require a data-driven dashboard that can help make informed decisions in content acquisition, marketing strategies, and production investments.`

Description:


A dynamic dashboard has been created by connecting MongoDB with Power BI to visually represent critical aspects of the movie dataset. The dashboard includes:

Total number of movies
Genre distribution
Top movies by IMDb rating
Language-wise and country-wise movie production
Movie production over years
Rating distribution analysis
This dashboard helps decision-makers easily monitor and interact with data for better business planning and strategic execution.

Objective of the Project:


Tools & Technologies Used:

MongoDB Compass for data querying and management.
CRUD Operations (Create, Read, Update, Delete) for modifying database records.
Data Visualization Dashboard for interactive analytics and graphical representation.

Steps Performed:


1. Data Ingestion
A sample movie database was imported into MongoDB Compass.
The database contained over 23,000 movie records, including fields like title, genre, year, writers, ratings, languages, etc.
2. CRUD Operations
Read Operation: Queried movies of specific genres,Retrieved movie documents matching the genre “Sci-Fi”.
Update Operation: Updated fields such as lastupdated, awards, and others.
Delete Operation: Deleted a document using _id.


 Observations:


English dominates the language of production, followed by a few regional or international languages.
The most common movie genre observed is Drama, followed by Comedy and Thriller.
Top movies with IMDb ratings above 8.5 tend to belong to Drama, Biography, or Crime genres.
Movie production peaked during specific years, showing trends in industry activity.
The number of movies per country shows a clear dominance of specific countries like the USA, UK, etc.
A clear distribution pattern is seen in ratings—most movies lie between 6.5 to 7.5 IMDb rating range.

Dashboard Visualizations &  Managerial Insights:

Dashboard Visuals and Managerial Insights
1. Total Number of Movies – 23,539
Insight:
A huge dataset of over 23K movies gives a robust base for analysis, ensuring more accurate trends and decisions.

2. Movie Ratings (Donut Chart)
Insight:
Majority of movies fall in the 6–7.5 rating range.
Very few movies exceed 8.5+ ratings, highlighting high-quality content scarcity.

3. Top 10 Genres by Number of Movies (Bar Chart)
Insight:
Drama leads with the highest number of movies, followed by Comedy, Thriller, and Action.
Indicates audience preference and production bias toward emotion-driven storytelling.

4. Average Metacritic Score (Gauge) – 59
Insight:
Indicates an average critical reception for most movies.
There is room to push for better content quality and script selection.

5. Top 10 Languages by Number of Movies (Bar Chart)
Insight:
English dominates, followed by Spanish, French, and German.
Suggests wider market reach and demand for English-language content.

6. Most Prolific Actors (Bar Chart - Horizontal)
Insight:
High engagement from stars like Robert De Niro, Samuel L. Jackson, and Bruce Willis.
Frequent appearances show their bankability and market pull.

7. Most Awarded Directors with Avg Metacritic Score (Table + Chart)
Insight:
Steven Spielberg and Martin Scorsese lead in awards and movie count.
Also show high critical scores, indicating a successful balance between quality and popularity.

8. Movies, Average Critics & Viewer Review Scores by Decade (Combo Chart)
Insight:
Movie production increased post-2000s, with a sharp rise in recent years.
However, average scores remain mostly stable, showing that quantity doesn’t always ensure quality.

9. Movie Runtime vs Metacritic Score (Heatmap)
Insight:
Optimal quality scores tend to cluster around 90–120 minutes runtime.
Very short or very long movies tend to have lower scores.

10. Average Metacritic Score by Country (Map Visualization)
Insight:
European and North American countries show better average scores.
Suggests better production quality and global reach in these regions.
